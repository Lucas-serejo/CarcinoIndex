# Extração de Características Geométricas a partir de Máscaras

**Objetivo:** Prototipar e analisar a extração de propriedades puramente geométricas (área em pixels, dimensões e *aspect ratio* da *bounding box*) a partir das máscaras binárias obtidas pela etapa de segmentação.

Essas *features* são projetadas para contornar a ambiguidade monocular de escala, mapeando relações espaciais relativas que servem de entrada para o modelo de classificação dos scores LS do Índice de Carcinomatose Peritoneal (PCI).

```python
import sys
from pathlib import Path

backend_dir = str(Path().resolve().parent / "backend")
if backend_dir not in sys.path:
    sys.path.append(backend_dir)

import numpy as np
import matplotlib.pyplot as plt
from ai.feature_extraction.geometric_features import extract_mask_area, extract_bounding_box_properties

%matplotlib inline
```

## 1. Simulação de uma Máscara de Lesão
Criamos uma máscara de teste para validar o comportamento das funções de extração.

```python
# Simula uma máscara binária de tamanho 200x200
mock_mask = np.zeros((200, 200), dtype=bool)
# Preenche uma região retangular representando um implante peritonial alongado
mock_mask[60:140, 80:110] = True

plt.figure(figsize=(4, 4))
plt.imshow(mock_mask, cmap="gray")
plt.title("Máscara Binária Simulada")
plt.axis("off")
plt.show()
```

## 2. Extração de Métricas
Invocamos o módulo especializado para quantificar a região.

```python
area = extract_mask_area(mock_mask)
bbox_props = extract_bounding_box_properties(mock_mask)

print(f"Área total da máscara: {area} pixels")
print(f"Propriedades da Bounding Box:")
print(f" - Coordenadas (x_min, y_min, x_max, y_max): {bbox_props['bbox']}")
print(f" - Largura: {bbox_props['width']}")
print(f" - Altura: {bbox_props['height']}")
print(f" - Aspect Ratio (L/A): {bbox_props['aspect_ratio']:.2f}")
```
